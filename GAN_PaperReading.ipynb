### Generative and Discriminative Model
- `Generative Model`
   - focus on *actual distribution* of individual class in dataset and learns the *joint probability* distribution
   - depends on *Bayes Theorem* to get conditional probability
   - generate more sample artificially
   - used more in unsupervised machine learning
   - Examples: Hidden Markov Model(HMM), LDA, GAN, Autoregression Model, Bayesian Network
- `Discriminative Model`
   - predict unseen data based on *conditional probability* directly
   - can be used for classification or regression tasks,more in supervised ml.
   - Examples: Logisitics regression, SVM, Traditional neural network, KNN, Decision Tree, Random Forest, CRF
####In Generative Adversarial Network:
- Generative model tries to produce fake currency, using it without detection.
- Discriminative model tries to detect the counterfeit currency. Namely, it detects whether a sample is from model/data distribution.1: $D(x)$, 0:$D(G(z))$

Both teams improve the model, until the fake currency and Indistinguishable.

Both are trained with **backpropagation and dropout**, and samples from G using *forward propagation*.

### Generative Adversarial Network
- Two-player minimax game with value function $V(G,D)$
截屏2022-12-31 10.04.34.png
- Notation:
   - $x$: data
   - $z$: noise sample
   - $p_g$: generator distribution
   - $p_z(z)$: input noise variable
   - $V(G,D)$: the degree of differences between real and generative sample.
   - $L=max_D V(G,D)$: fixed generator G, let discriminator classify whether the data from actual or generative sample.
   - $min_G L$: fixed discriminator D, let generator G to realize the minimum differences between sample generated and realdata.
   - $G(z;θ_g)$: differentiable function represented by a `multilayer perception(MLP)`
      - **MLP**: 
        - a supplement of FFNN(feed forward neural network). 
        - A type of neural network that is composed of multiple layers of interconnected “neurons”
        - adjust the weights of the connections between neurons to minimize the error between the predicted output and the true ouput.
        - new input data pass through *forward propagation*, and produce the predicted output. 
   - $D(x;θ_d)$: multiplayer perception, represent the probability $x$ comes from the **data** rather than $p_g$.
- Training:
   - $D$
     - Train $D$ to <u>maximize</u> the probability of <u>assigning correct label</u> to training samples and G. 
     - Optimize $D$ in a inner loop of training → overfitting and computational costy
   - $G$
     - Train $G$ to <u>minimize</u> $log(1-D(G(z)))$ or <u>maximize</u> $logD(G(z))$. The second option provides much stronger gradients early in learning.
   - `General Method`: Alternate between k steps of optimizing $D$ and one step of optimizing $G$.

   - The global optimum: $p_g=p_x$

### Evaluation
#### Advantage
   - Inputs are not directly copied into the generator's parameter
   - can present very sharp, degenerate distributions. Any differentiable function is permitted.
#### Disadvantage
   - Unstable training
   - evaluating $p_g$ is subjective.

### Application:
- Text-to-image, Image-to-text
- Style Transfer
- Data Augmentation
- Image Super Resolution
